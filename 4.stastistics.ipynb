```python
# Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
# Distribution and Binomial Distribution?



# The Bernoulli Distribution is a probability distribution that represents
# a random experiment with two possible outcomes: success and failure. 
# It's named after the Swiss mathematician Jacob Bernoulli, who introduced
# it in the 18th century. The distribution is characterized by a single 
# parameter, usually denoted as "p," which represents the probability of
# success (the probability of an event occurring).


# Example:
# Consider a coin toss, where we define "heads" as a success (X = 1) and
# "tails" as a failure (X = 0). If the coin is fair, the probability of
# getting heads (success) is p = 0.5, and the probability of getting tails 
# (failure) is 1 - p = 0.5. This scenario can be modeled using a Bernoulli 
# distribution.


# Bernoulli Distribution:

# Represents a single trial or experiment with two possible outcomes (success and failure).
# Has one parameter, p, which represents the probability of success in a single trial.
# Denoted as B(p), where B represents the Bernoulli Distribution.
# Binomial Distribution:

# Represents the number of successes (k) in a fixed number (n) of independent Bernoulli trials.
# Has two parameters: p (the probability of success in a single trial) and n (the number of trials).
# Denoted as Binomial(n, p).
```

```python
# Q7: Explain uniform Distribution with an example.


# The Uniform Distribution is a probability distribution that describes
# a situation where all values in a given range are equally likely to 
# occur. In other words, every possible outcome within the range has the
# same probability of occurring. This distribution is often denoted as
# U(a, b), where "a" and "b" are the parameters representing the lower
# and upper bounds of the range, respectively.

# Example:
# Imagine you have a fair six-sided die. When you roll the die, the 
# outcome represents a random variable that follows a discrete 
# Uniform Distribution, denoted as U(1, 6). In this case:

# "a" is the lower bound of the range, which is 1 (the minimum value on the die).
# "b" is the upper bound of the range, which is 6 (the maximum value on the die).
```

```python
# Q8: What is the z score? State the importance of the z score.

# The z-score, also known as the standard score or standard deviation score,
# is a statistical measure that quantifies how many standard deviations
# a data point is away from the mean (average) of a dataset. 
# It is calculated using the following formula:

# z = (X - μ) / σ

# z is the z-score.
# X is the individual data point.
# μ (mu) is the mean (average) of the dataset.
# σ (sigma) is the standard deviation of the dataset.

# The z-score is important for several reasons:

# Finance and Risk Management: In finance, z-scores can be used to assess
# the creditworthiness and financial health of individuals or companies. 
# They are also used in risk management to evaluate asset performance.
```

```python
# Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

# The Central Limit Theorem (CLT) is a fundamental principle in probability
# theory and statistics that describes the behavior of the sampling 
# distribution of the sample mean (or sum) as the sample size increases, 
# regardless of the shape of the original population distribution.
# It states that, under certain conditions, the sampling distribution
# of the sample mean approaches a normal distribution as the sample 
# size grows, irrespective of the population distribution.


# Foundation of Hypothesis Testing: The CLT is the foundation for many
# hypothesis tests involving the sample mean, enabling researchers to 
# draw conclusions about the population mean based on sample data.
```
